```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_medidas = pd.DataFrame.from_dict(dict_faturamento)

df_medidas
```

```python
## Calcular a média dos valores 
df_medidas['valor'].mean()
```

```python
## Grafico de barras verticais Dos meses com maior faturamento
df_medidas.plot.bar(x='data_ref', y='valor', title='Faturamento Mensal', figsize=(10,5))
```

```python
## Grafico de Linhas mostrando o mês de referência e o valor.
df_medidas.plot.line(x='data_ref', y='valor', title='Faturamento Mensal', figsize=(10,5))
```
